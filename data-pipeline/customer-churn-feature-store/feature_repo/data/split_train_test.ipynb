```python
import pandas as pd
from pathlib import Path

data_dir = Path(__file__).parent if '__file__' in dir() else Path().resolve()

df = pd.read_parquet(data_dir / 'processed_churn_data.parquet')
```

```python
train_df = df.sample(frac=0.8, random_state=100)
test_df = df.drop(train_df.index)
```

```python
train_df.to_parquet(data_dir / 'train.parquet', index=False)
test_df.to_parquet(data_dir / 'test.parquet', index=False)

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
```
